```python
import pandas as pd
```

```python
database1 = pd.read_csv("cancerdata.csv")
database1
```

```python
database1.info()
```

```python
database1.drop('id',axis=1,inplace=True)
database1.drop('Unnamed: 32',axis=1,inplace=True)
database1.info()
```

```python
# Quantos dados de cada Especie
database1['diagnosis'].count()
```

```python
database1['diagnosis'].value_counts()
```

```python
database1.diagnosis.unique()
```

```python
database1['diagnosis'] = database1['diagnosis'].map({'M':1,'B':0})
database1.head()
```

```python
correlacao = database1.corr()
correlacao
```

```python
X = database1.drop('diagnosis',axis=1).values
y = database1['diagnosis'].values
```

```python
X
```

```python
y
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.20)
print(len(x_treinamento),len(x_teste))
```

```python
#Quando C=1, representa um equilíbrio entre maximizar a margem e minimizar o erro de classificação
classificador = svm.SVC(kernel='linear', C=1.0)
```

```python
classificador.fit(x_treinamento, y_treinamento)
```

```python
y_teste
```

```python
y_preditivo = classificador.predict(x_teste)
y_preditivo
```

```python
#Acurácia
classificador.score(x_teste,y_teste)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_teste, y_preditivo)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_teste, y_preditivo))
```
