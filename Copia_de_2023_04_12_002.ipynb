```python
X = database1.drop('Wine',axis=1).values
y = database1['Wine'].values
```

```python
X
```

```python
y
```

```python
# Importar as bibliotecas necessárias
from sklearn import svm 
from sklearn.model_selection import train_test_split
```

```python
# Dividir os dados em conjuntos de treinamento e teste:
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.20, random_state=0)
print(len(x_treinamento),len(x_teste))
```

```python
# Treinar o modelo SVM com os dados de treinamento
classificador = svm.SVC(kernel='linear', C=1.0)
classificador.fit(x_treinamento, y_treinamento)
```

```python
y_teste
```

```python
# Fazer previsões usando o modelo treinado e os dados de teste
y_preditivo = classificador.predict(x_teste)
y_preditivo
```

```python
#Acurácia
classificador.score(x_teste,y_teste)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix # Importar as bibliotecas necessárias
confusion_matrix(y_teste, y_preditivo)
accuracy_score(y_teste, y_preditivo)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_teste, y_preditivo))
```

```python
# Avaliar o desempenho do modelo usando várias métricas, como acurácia, matriz de confusão e relatório de classificação
print('Acurácia:', accuracy_score(y_teste, y_preditivo))
print('Matriz de confusão:\n', confusion_matrix(y_teste, y_preditivo))
print('Relatório de classificação:\n', classification_report(y_teste, y_preditivo))
```
